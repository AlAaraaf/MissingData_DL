## Generate NHANES Continuous dataset

```python
import pandas as pd
import numpy as np

# data folder path
data_path = "../dataset/NHANES/"

# Read in datasets and drop rows with missing data
# demo
demo_col = ["SEQN", "FIALANG", "RIDEXMON", "DMQMILIZ", "DMDBORN4", "DMDCITZN", "SIAPROXY"]
demo1 = pd.read_sas(data_path + "DEMO_J.XPT", format="xport", encoding='utf-8')
demo1 = demo1[demo_col]
demo2 = pd.read_sas(data_path + "DEMO_I.XPT", format="xport", encoding='utf-8')
demo2 = demo2[demo_col]

# body measure
bm_col = ["SEQN","BMXWT", "BMXHT"]
bm1 = pd.read_sas(data_path + "BMX_J.XPT", format="xport", encoding='utf-8')
bm1 = bm1[bm_col]
bm2 = pd.read_sas(data_path + "BMX_I.XPT", format="xport", encoding='utf-8')
bm2 = bm2[bm_col]

# blood pressure
bp_col = ["SEQN","BPXPLS", "BPXML1", "BPXSY1", "BPXDI1"]
bp1 = pd.read_sas(data_path + "BPX_J.XPT", format="xport", encoding='utf-8')
bp1 = bp1[bp_col]
bp2 = pd.read_sas(data_path + "BPX_I.XPT", format="xport", encoding='utf-8')
bp2 = bp2[bp_col]

# consumer
cons_col = ["SEQN","CBD071", "CBD091", "CBD121"]
cons1 = pd.read_sas(data_path + "CBQ_J.XPT", format="xport", encoding='utf-8')
cons1 = cons1[cons_col]
cons2 = pd.read_sas(data_path + "CBQ_I.XPT", format="xport", encoding='utf-8')
cons2 = cons2[cons_col]

# health insurance
hi_col = ["SEQN", "HIQ011"]
hiq1 = pd.read_sas(data_path + "HIQ_J.XPT", format="xport", encoding='utf-8')
hiq1 = hiq1[hi_col]
hiq2 = pd.read_sas(data_path + "HIQ_I.XPT", format="xport", encoding='utf-8')
hiq2 = hiq2[hi_col]

# income
inc_col = ["SEQN","INQ020","INDFMMPI"]
inc1 = pd.read_sas(data_path + "INQ_J.XPT", format="xport", encoding='utf-8')
inc1= inc1[inc_col]
inc2 = pd.read_sas(data_path + "INQ_I.XPT", format="xport", encoding='utf-8')
inc2= inc2[inc_col]

# disability
dis_col = ["SEQN", "DLQ040", "DLQ050"]
dis1 = pd.read_sas(data_path + "DLQ_J.XPT", format="xport", encoding='utf-8')
dis1 = dis1[dis_col]
dis2 = pd.read_sas(data_path + "DLQ_I.XPT", format="xport", encoding='utf-8')
dis2 = dis2[dis_col]

# Merge datasets
data1 = demo1.merge(bm1, on="SEQN").merge(bp1, on="SEQN").merge(cons1, on="SEQN").merge(hiq1, on="SEQN").merge(inc1, on="SEQN").merge(dis1, on="SEQN")
data1 = data1.dropna()
data2 = demo2.merge(bm2, on="SEQN").merge(bp2, on="SEQN").merge(cons2, on="SEQN").merge(hiq2, on="SEQN").merge(inc2, on="SEQN").merge(dis2, on="SEQN")
data2 = data2.dropna()

data = data1.append(data2, ignore_index = True)
data.head()
```

```python
# data cleaning

# remove data with CBD071/091/121 = 999999,777777
data_new1 = data[~data['CBD071'].isin([999999, 777777,8400])]
data_new2 = data_new1[~data_new1['CBD091'].isin([999999, 777777])]
data_new3 = data_new2[~data_new2['CBD121'].isin([999999, 777777,8400])]
data_new = data_new3.replace(5.397605346934028e-79, 0)
```

```python
# Calculate categorical variables' levels
categorical_columns = ['FIALANG', 'RIDEXMON', 'DMQMILIZ', 'DMDBORN4', 'INQ020',
                       'DMDCITZN', 'SIAPROXY', 'HIQ011', 'DLQ040', 'DLQ050']
print("Unique counts for categorical variables:")
for column in categorical_columns:
    print("{} - {}:({},{})".format(column, 
                                   np.unique(data_new[column]), 
                                   np.min(data_new[column]), 
                                   np.max(data_new[column])))

# Calculate the range for continuous variables
continuous_columns = [x for x in list(data_new.columns) if x not in categorical_columns]
for column in continuous_columns:
    print("{} - ({},{})".format(column, 
                                np.min(data_new[column]), 
                                np.max(data_new[column])))
```

```python
# re-labeling
from scipy.stats import rankdata
for column in categorical_columns:
    data_new[column] = (rankdata(data_new[column], method = 'dense') - 1).astype(int)


print("Unique counts for categorical variables:")
for column in categorical_columns:
    print("{} - {}:({},{})".format(column, 
                                   np.unique(data_new[column]), 
                                   np.min(data_new[column]), 
                                   np.max(data_new[column])))
```

```python
# remove ID column
# sort dataset: [categorical variables, continuous variables]
continuous_columns.remove('SEQN')
sort_columns = categorical_columns.copy()
sort_columns.extend(continuous_columns)
data_new = data_new[sort_columns]
data_new.head()
```

```python
print("Number of categorical variables: {}".format(len(categorical_columns)))
print("Number of continuous variables: {}".format(len(continuous_columns)))
```

```python
data_new.to_csv('../dataset/nhanes.csv',index=False)
```
